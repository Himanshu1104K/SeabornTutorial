```python
import pandas as pd 
import seaborn as sns 
import matplotlib.pyplot as plt
```

```python
tips = sns.load_dataset('tips')
tips.head()
```

```python
sns.regplot(x='total_bill',y='tip' ,data=tips)
```

```python
sns.regplot(x='total_bill',y='tip',data=tips,ci=None)
```

```python
sns.regplot(x='total_bill',y='tip',data=tips,scatter=None)
```

```python
sns.lmplot(x='total_bill',y='tip',hue='sex',data=tips)
```

```python
sns.lmplot(x='total_bill',y='tip',hue='sex',data=tips,ci=None)
```

```python
sns.lmplot(x='total_bill',y='tip',hue='sex',data=tips,scatter=False)
```

```python
iris = sns.load_dataset('iris')
iris.head()
```

```python
g = sns.PairGrid(iris)
g.map(plt.scatter,marker='.')
```

```python
g = sns.PairGrid(iris)
g.map_upper(sns.scatterplot)
g.map_lower(sns.kdeplot)
g.map_diag(sns.histplot)
```

```python
sns.set_context("talk")
sns.boxplot(x="day", y="total_bill", data=tips)
plt.show()
```

```python
sns.set_context("talk")
sns.boxplot(x="day", y="total_bill", data=tips)
sns.despine()
plt.show()
```
