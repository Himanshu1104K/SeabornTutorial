```python
import pandas as pd 
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

```python
sns.barplot(x='day',y='total_bill',data=tips)
plt.title("Average Total Bill per Day")
```

```python
sns.countplot(x='day',data=tips)
plt.title("Number of Meals per Day")
plt.tight_layout()
```

```python
sns.barplot(x="day", y="total_bill", hue="sex", data=tips)
plt.title("Average Total Bill per Day by Gender")
```

Practise Questions 

```python
# Create a bar plot showing the average tip amount for each day

sns.barplot(x='day',y='tip',data=tips)
plt.title("Average Tip Amount Per day")
```

```python
sns.barplot(x='day',y='tip',hue='sex',data=tips)
plt.title("Average Tip Amount Per day")
```

```python
# Plot the count of meals served during lunch and dinner times 

sns.countplot(x='time',data=tips)
```

```python
# Plot the distribution of tips for each day using both box and violin plots.

sns.boxplot(x='day',y='tip',data=tips)
```

```python
sns.violinplot(x='day',y='tip',data=tips)
```

```python
# Use hue to differentiate between smokers and non-smokers.

sns.boxplot(x='day',y='tip',hue="smoker",data=tips)
```

```python
sns.violinplot(x='day',y='tip',hue="smoker",data=tips)
```

```python
# Create a strip plot of total bill by day, adding jitter to spread out the points.

sns.stripplot(x='day',y='total_bill',data=tips,jitter=0.25)
```

```python
sns.violinplot(x='day',y='total_bill',data=tips,inner=None)
sns.swarmplot(x='day',y='total_bill',data=tips,color="white",edgecolor="auto")
```
